```python
import cv2
import numpy as np
from wrapper_vpgnet import WrapperVPGNet
import utils
from matplotlib import pyplot as plt

model_path = '/home/master/09/luben3485/csie5452-lane-detection/vpgnet_weights/vpgnet_146_455_01.pth'
#model_path = '/home/master/09/luben3485/csie5452-lane-detection/vpgnet_weights_136_8_50_50/vpgnet_epoch_5.pth'

vpgnet = WrapperVPGNet(model_path)
#image_origin = cv2.imread('rgb_img.jpg', cv2.IMREAD_COLOR)
image_origin = cv2.imread('/tmp2/chacotw/car/csie5452-lane-detection/vpgnet-pytorch/testcrop.jpg', cv2.IMREAD_COLOR)
print(image_origin.shape)

copy_image_origin_vpgnet = cv2.resize(image_origin.copy(), (640, 480))
obj_mask_pred_160x120_vpgnet = vpgnet.get_lane_line_from_image_640x480(copy_image_origin_vpgnet)
#print(obj_mask_pred_160x120_vpgnet.shape)


# obj_mask_pred_1920x1080_vpgnet = np.resize(obj_mask_pred_160x120_vpgnet, (4, 1080, 1920))

#lane_pred = utils.resize_array(obj_mask_pred_160x120_vpgnet[0, :, :], (939, 1921))
lane_solid_white = utils.resize_array(obj_mask_pred_160x120_vpgnet[0, :, :], (222, 455))
lane_solid_yellow = utils.resize_array(obj_mask_pred_160x120_vpgnet[1, :, :], (222, 455))
lane_double_yellow = utils.resize_array(obj_mask_pred_160x120_vpgnet[2, :, :], (222, 455))

threshold = 0.7
lane_solid_white = np.where(lane_solid_white > threshold, 1, 0)
lane_solid_yellow = np.where(lane_solid_yellow > threshold, 1, 0)
lane_double_yellow = np.where(lane_double_yellow > threshold, 1, 0)

'''
mask = cv2.imread('mask.jpg')
#print(mask.shape)
#print(mask.nonzero())
plt.figure(figsize=(10,5))
plt.axis('off')
plt.imshow(mask)
plt.show()
'''

'''
plt.figure(figsize=(10,5))
plt.axis('off')
plt.imshow(image_origin[:,:,::-1])
plt.show()
'''


color_map_mat = np.zeros((19, 3), dtype=np.uint8)
for i in range(0, 19):
    if i == 1:
        # lane_solid_white
        color_map_mat[i] = (0, 0, 255)
    elif i == 4:
        # lane_solid_yellow
        color_map_mat[i] = (0, 0, 255)
    elif i == 6:
        #lane_double_yellow
        color_map_mat[i] = (255, 0, 0)
    elif i == 18:
        # erfnet_mask
        color_map_mat[i] = (0, 0, 255)

prob_map_lane_solid_white = lane_solid_white * 1
prob_map_lane_solid_yellow = lane_solid_yellow * 4
prob_map_lane_double_yellow = lane_double_yellow * 6

seg_image_lane_solid_white = color_map_mat[prob_map_lane_solid_white]
seg_image_lane_solid_yellow = color_map_mat[prob_map_lane_solid_yellow]
seg_image_lane_double_yellow = color_map_mat[prob_map_lane_double_yellow]

gamma = 0
img_lane_all = cv2.addWeighted(seg_image_lane_solid_yellow, 1, seg_image_lane_solid_white, 1, gamma)
img_lane_all = cv2.addWeighted(seg_image_lane_double_yellow, 1, img_lane_all, 1, gamma)
img_lane_all = cv2.blur(img_lane_all, (9, 9))
img_lane_all = cv2.addWeighted(image_origin[:,:,::-1], 0.8, img_lane_all, 1, gamma)

plt.figure(figsize=(10,5))
plt.axis('off')
plt.imshow(img_lane_all)
plt.show()


print('lane_solid_white')
plt.figure(figsize=(10,5))
plt.axis('off')
plt.imshow(lane_solid_white, cmap='gray')
plt.show()

print('lane_solid_yellow')
plt.figure(figsize=(10,5))
plt.axis('off')
plt.imshow(lane_solid_yellow, cmap='gray')
plt.show()

print('lane_double_yellow')
plt.figure(figsize=(10,5))
plt.axis('off')
plt.imshow(lane_double_yellow, cmap='gray')
plt.show()
```

```python
import torch
from torch import nn
m = nn.Sigmoid()
loss = nn.BCELoss()
input = torch.randn(3,3, requires_grad=True)
print(input)
target = torch.empty(3,3).random_(2)
print(target)

a = loss(m(input[0]), target[0])
b = loss(m(input[1]), target[1])
c = loss(m(input[2]), target[2])

print(a,b,c)
print((a+b+c)/3)
output = loss(m(input), target)
print(output)
output.backward()
```
